**Objetivo**: Construir uma análise completa do comportamento de vendas de uma empresa, usando Python para processamento de dados e Power BI para visualização.

## Passo 1: Coleta e Preparação dos Dados

Dataset: Use um dataset de vendas que contenha informações sobre data, produto, quantidade vendida, valor, localização e cliente (você pode encontrar no Kaggle ou criar dados fictícios).
Python: Utilize bibliotecas como pandas para ler e manipular os dados. Corrija dados nulos, ajuste formatos de data e crie novas colunas, como "receita total" (quantidade × preço).

```python
import pandas as pd
import numpy as np

# Carregando Dataframe
df = pd.read_excel('/content/drive/MyDrive/Documentos/Cursos e Projetos/Projeto - Análise de Vendas/Vendas.xlsx')
df
```

### Criando nova coluna

```python
# Dados da nova coluna
nomes = ['Pedro', 'Ana', 'João','Miguel', 'Maria']

# Cria a nova coluna repetindo os nomes até o tamanho do DataFrame
df['Cliente'] = np.tile(nomes, len(df) // len(nomes) + 1)[:len(df)]

# Move a coluna para o primeiro índice
df.insert(0, 'Cliente', df.pop('Cliente'))

df
```

### Limpando dados

```python
# Remove as linhas com produto "Bermuda"
df = df[df['Produto'] != 'Bermuda']
df
```

```python
# Reseta o índice das linhas
df = df.reset_index(drop=True)
df
```

## Passo 2: Análise Exploratória dos Dados (EDA)
**Python**: Use `matplotlib` e `seaborn` para explorar os dados e visualizar tendências. Crie gráficos de distribuição das vendas ao longo do tempo, categorias de produtos mais vendidos e análise de vendas por região.

### Matplotlib

```python
import matplotlib.pyplot as plt

# Define o tamanho da figura
fig, ax = plt.subplots(figsize=(15, 7))

# Agrupa os dados por produto e soma as quantidades
df_grouped = df.groupby('Produto')['Quantidade'].sum().reset_index()

# Cria o gráfico de barras
barras = ax.bar(df_grouped['Produto'], df_grouped['Quantidade'])

# Adiciona rótulos e título
ax.set_xlabel("Quantidade vendida")
ax.set_ylabel("Produto")
ax.set_title("Produto x Quantidade")

# Adiciona os valores ao lado de cada barra
ax.bar_label(barras, padding=3)

plt.show()
```
